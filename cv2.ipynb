##함수의 오류를 방지하기 위해 cv2_imshow import


```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

##이미지 경로설정 및 imread() 함수를 이용해 저장된 이미지의 경로 불러오기

```python
image_path ='/content/cat.jpg'
img = cv2.imread(image_path)
```

##이미지를 그레이스케일로 변환

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

 ## 가장자리 검출을 위해 적절한 커널을 사용한 블러링

```python
blurred = cv2.GaussianBlur(gray, (5, 5), 0)
```

## 캐니 엣지 검출

```python
edges = cv2.Canny(blurred, 50, 150)
```

## 검출된 가장자리를 컬러 이미지로 변환하여 표시

```python
edges_colored = cv2.cvtColor(edges, cv2.COLOR_GRAY2BGR)
```

## 결과 이미지를 화면에 표시


```python
cv2_imshow(img)
cv2_imshow(edges_colored)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
